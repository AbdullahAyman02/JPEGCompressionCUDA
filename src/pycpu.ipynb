```python
from pycpu import JPEGCompression

# Example usage
compressor = JPEGCompression(quality=75)  # Adjust quality as needed
image = compressor.read_image("./dump/R.jpg")
compressed_image = compressor.compress_image(image)
compressor.display_image(image, "Original Image")
compressor.display_image(compressed_image, "Compressed Image")
```

```python
compressor.save_image(compressed_image, "./dump/" + "compressed.jpg")  # Save the compressed image

# Calculate and print compression ratio
original_size = image.nbytes / 1024  # Size in KB
compressed_size = compressed_image.nbytes / 1024 # Size in KB (Note: this is in-memory size, file size might differ due to further encoding in imwrite)
compression_ratio = original_size / compressed_size
print(f"Compression Ratio: {compression_ratio:.2f}")

import numpy as np
# Calculate and print MSE and PSNR
mse = np.mean((image - compressed_image) ** 2)
psnr = 10 * np.log10((255 ** 2) / mse) if mse > 0 else float('inf') # Handle potential division by zero
print(f"MSE: {mse:.2f}")
print(f"PSNR: {psnr:.2f} dB")
```
